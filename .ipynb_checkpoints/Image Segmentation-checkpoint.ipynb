```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
im = cv2.imread('Apoorv.jpeg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
original_shape = im.shape
print(original_shape)
```

```python
plt.style.use('seaborn')
```

```python
plt.imshow(im)
plt.axis('off')
plt.plot()
```

### Now we will flatten the array

```python
all_pixels = im.reshape(original_shape[0] * original_shape[1], 3)
```

```python
all_pixels.shape
```

```python
dominant_colors = 6
km = KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)
```

```python
centers = km.cluster_centers_
```

```python
centers
```

```python
centers = np.array(centers, dtype='uint8')
```

```python
centers
```

```python
### Just to check
for c in centers:
    print(c)
```

#### Plot what all these are

```python
i = 1

plt.figure(figsize=(10, 3))

for each_color in centers:
    plt.subplot(1, dominant_colors, i)
    plt.axis('off')
    i += 1
    
    a = np.zeros((100, 100, 3), dtype='uint8')
    a[:, :, :] = each_color
    plt.imshow(a)

plt.show()
    
```

```python
predictions = km.labels_
print(predictions[0:20])
```

```python
type(predictions)
```

```python
predictions.shape
```

```python
new_img = np.zeros((original_shape[0] * original_shape[1], 3), dtype='uint8')

for i in range(predictions.shape[0]):
    new_img[i] = centers[predictions[i]]

new_img = new_img.reshape(original_shape)

    
```

```python
plt.imshow(new_img)
plt.axis('off')
plt.show()
```
